```python
from google_play_scraper import Sort, reviews_all

import numpy as np
```

```python
np.random.randint(500, 2000)
```

```python
result = reviews_all(
    'com.miHoYo.GenshinImpact',
    sleep_milliseconds=np.random.randint(500, 2000), # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
)

len(result)
```

```python
all_reviews_most_rel = []

for i in range(10):
    result = reviews_all(
        'com.miHoYo.GenshinImpact',
        sleep_milliseconds=np.random.randint(500, 2000),  # defaults to 0
        lang='en',  # defaults to 'en'
        country='us',  # defaults to 'us'
        sort=Sort.MOST_RELEVANT,  # defaults to Sort.MOST_RELEVANT
    )
    
    all_reviews_most_rel += result

    len(all_reviews_most_rel)
```
